### Subtract Min
Given a list X of ints, create a list Y where
\begin{align}
    Y[i] = X[i] - X_{min}
\end{align}
where $X_{min}$ is the minimum value in list X 

```python
def subtract_min(X):
    """
    Input:
        X : list of ints
    Output :
        Y : list of ints, 
        each element of Y is the corresponding element of X subtracted by minimum value in list X
    """
    # YOUR CODE HERE
    X=[1,2,3,4,5,6,7,8]
    la=np.array(X)
    Y=[]
    j=la.min()
    for i in X:
        la = la - [j]
    Y.append(la)
    
        
         
    
    return Y
```
